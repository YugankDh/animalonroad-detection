```python
from ultralytics import YOLO

model = YOLO("yolo11m.pt")

model.train(source='/content/animals-on-road-3/data.yaml')
```

```python
import cv2

model = YOLO("/content/runs/detect/train/weights/best.pt")

image = cv2.VideoCapture("/content/input.mp4")

animal_labels = [
    'bird', 'cat', 'dog', 'horse', 'sheep', 'cow', 'elephant', 'bear', 'zebra', 'giraffe'
]

fps = image.get(cv2.CAP_PROP_FPS)
frame_height = int(image.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_width = int(image.get(cv2.CAP_PROP_FRAME_WIDTH))
out = cv2.VideoWriter("output.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (frame_width, frame_height))

while True:
  ret, frame = image.read()

  if not ret:
    break

  animal_detected_in_frame = False
  results = model.predict(source=frame, verbose=False)

  for result in results:
    for box in result.boxes:
      label = model.names[int(box.cls)]

      if label in animal_labels:
        animal_detected_in_frame = True
        x1, y1, x2, y2 = map(int,box.xyxy[0])

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

  if animal_detected_in_frame:
      cv2.putText(frame, "WARNING - Animal ahead!", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2) # Top-left warning

  out.write(frame)

image.release()
out.release()
```
